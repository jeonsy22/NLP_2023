NaN (NaN-Data; 결측값)
* Not a Number: 비어있는 셀 값
* 배열에서 연살할 경우 오류가 발생하지 않는다.
* 결과값으로 NaN을 출력해준다.
* foat형에서만 작동함

```python
import numpy as np
import pandas as pd

x01 = np.random.randint(50, 100, size = (4, 3))
x01
```

```python
x02 = pd.DataFrame(x01, index = ['1반', '2반', '3반', '4반'], columns = ['국어', '수학', '영어'])
x02
```

```python
y01 = x02.astype('float')
y01
```

```python
y01['영어'][2] = np.nan
y01
```

```python
y02 = y01.dropna(axis = 0)
y02
```

```python
y03 = y01.dropna(axis = 1)
y03
```

```python
y04 = y01.fillna('1000')
y04
```

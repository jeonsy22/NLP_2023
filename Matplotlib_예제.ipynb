```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [69, 57, 88, 79, 80]
y2 = [78, 92, 77, 69, 94]
plt.figure(figsize = (5, 5)) #그래프 사이즈 지정
plt.plot(x, y1, color = 'g', label = 'math')
plt.bar(x, y2, color = 'b', label = 'english')
plt.xlabel('number')
plt.ylabel('score')
plt.legend()
plt.title('Score calculation for each subject', fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [69, 57, 88, 79, 80]
y2 = [78, 92, 77, 69, 94]
plt.figure(figsize = (10, 5)) #그래프 사이즈 지정
plt.plot(x, y1, color = 'g', label = 'math')
plt.bar(x, y2, color = 'b', label = 'english')
plt.xlabel('number')
plt.ylabel('score')
plt.legend()
plt.title('Score calculation for each subject', fontsize = 30)
plt.xticks(rotation = 45)
plt.show()
```

```python
import numpy as np
import pandas as pd

x01 = np.random.randint(50, 100, size = (4, 3))
x01
```

```python
x02 = pd.DataFrame(x01, index = ['1반', '2반', '3반', '4반'], columns = ['국어', '수학', '영어'])
x02
```

```python
x02['국어']
```

```python
x02.loc['2반']
```

```python
x02.iloc[3]
```

```python
x02['평균'] = round((x02['국어'] + x02['영어'] + x02['수학'])/3, 2)
x02
```

```python
x02['합계'] = np.nan
x02
```

```python
del x02['합계']
x02
```

```python
import numpy as np
import pandas as pd

x01 = np.random.randint(50, 100, size = (4, 3))
x01
```

```python
x02 = pd.DataFrame(x01, index = ['1반', '2반', '3반', '4반'], columns = ['국어', '수학', '영어'])
x02
```

```python
x02['과학'] = [80, 90, 53, 72]
x02
```

```python
del x02['과학']
x02
```

```python
x02.insert(2, '과학', [80, 90, 53, 72])
x02
```

```python
x03 = x02.drop('3반', axis = 'rows')
x03
```

```python
x03 = x02.drop('3반', axis = 0)
x03
```

```python
x04 = x03.drop('과학', axis = 'columns')
x04
```

```python
x05 = x04.drop('수학', axis = 1)
x05
```

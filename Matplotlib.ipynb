#Matplotlib 기본 활용

### 그래프 기본 명령어
* 그래프 그리기 모듈 호출
  * import matplotlib.pyplot as plt
* plot: '가상의 공간에 그래프를 그려라' 의미
* plt.show(): '완성된 그래프를 출력하라' 의미
* 그래프 사이즈: plt.figure(figsize = x, y)) #x, y값 지정
* x축 제목: plt.xlabel('name') #name 값 지정
* y축 제목: plt.ylabel('name') #name 값 지정
* 범례: plt.legand()
* 그래프제목: plt.title('name', fontsize = 30) #name 값 지정, 30: 글꼴 크기
* 그래프 출력하기: plt.show()

plot (꺽은 선 그래프)

```python
import matplotlib.pyplot as plt
plt.plot([1,2, 3, 4])
plt.show()
```

plt.plot([1, 2, 3, 4]): 기본적으로 plot(x, y) 순서로 2개의 리스트를 넣어주지만, 만약 위와 같이 하나의 리스트만 넣을 경우 x축 값은 자동으로 0, 1, 2, 3이고, plot([1, 2, 3, 4])의 값들은 y축 값이 된다.

bar (막대그래프)

```python
p = (['1', '2', '3', '4'])
q = [89, 78, 90, 94]
plt.bar(p, q)
plt.show()
```

scatter (산점도그래프)

```python
x = [30, 56, 73, 67, 49]
y = [3, 8 ,14, 7, 5]
plt.scatter(x, y)
plt.show()
```
